```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
data = {
    'Tinggi': [150, 155, 160, 165, 170, 175],
    'Berat_Badan': [55, 60, 65, 70, 75, 80]
}

df = pd.DataFrame(data)
```

```python
x = df[['Tinggi']]
y = df[['Berat_Badan']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred)
plt.title('Regresi Linear_Tingi dan Berat Badan')
plt.xlabel('Tinggi_cm')
plt.ylabel('Berat Badan_kg')
plt.show()
```

```python
tinggi_baru = [[158]]
berat_badan_prediksi = model.predict(tinggi_baru)

print(f'Tinggi: {tinggi_baru[0][0]} cm')
print(f'Berat Badan Prediksi: {berat_badan_prediksi[0]}kg')
```
